```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
import pandas as pd
```

```python
data=pd.read_csv('C://Users/Hp/Desktop/Data Science Assignment/Association Rules/book.csv')
```

```python
data
```

## apriori algorithm

```python
frequent_itemset=apriori(data,min_support=0.1,use_colnames=True)
```

```python
frequent_itemset
```

```python
rules=association_rules(frequent_itemset,metric='lift',min_threshold=0)
```

```python
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
rules[['support','confidence','lift']].hist(figsize=(15,7))
plt.show()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.scatterplot('support','confidence',data=rules)
plt.show()
```

```python
import pandas as pd
```

```python
data1=pd.read_csv('C://Users/Hp/Desktop/Data Science Assignment/Association Rules/my_movies.csv')
```

```python
data1
```

```python
data2=data1[['Sixth Sense','Gladiator','LOTR1','Harry Potter1','Patriot','LOTR2','Harry Potter2','LOTR','Braveheart','Green Mile']]
```

```python
data2
```

## APRIORI AGLORITHM

```python
frequent_itemsets=apriori(data2,min_support=0.1,use_colnames=True)
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
Rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
```

```python
Rules
```

```python
Rules.sort_values('lift',ascending=False)
```

```python
import matplotlib.pyplot as plt
Rules[['support','confidence','lift']].hist(figsize=(15,7))
plt.show()
```

```python
plt.scatter(Rules['support'],Rules['confidence'])
plt.show()
```

```python
import seaborn as sns
sns.scatterplot('support','confidence',data=Rules)
plt.show()
```
